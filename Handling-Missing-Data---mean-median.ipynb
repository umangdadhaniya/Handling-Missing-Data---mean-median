```python
missing_val_per = df.isnull().sum()/df.shape[0] * 100
missing_val_per
```

```python
missing_val_column_20 = missing_val_per[missing_val_per > 20].keys()
missing_val_column_20
```

```python
df2_drop_column = df.drop(columns=missing_val_column_20)
df2_drop_column.shape
```

```python
df3_num = df2_drop_column.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
```
